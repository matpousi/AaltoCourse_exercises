# Static game of incomplete information with multiple equilibria

This is an example of a static game of incomplete information with multiple equilibria. The model is described in detail in the lecture notes. Here we illustrate how to use the `sgame` class to solve for equilibria, plot best response functions, simulate data, and compute the log-likelihood

```python
from sgame import *
%load_ext autoreload
%autoreload 2
```

```python
model = sgame(alpha=5, beta=-11, x_a=0.52, x_b=0.18)

pa=model.solve()
pb=model.br_b(pa)
print(f'Eqb CCPs firm a: [{pa}]')
print(f'Eqb CCPs firm b: [{pb}]')
# Plot best response functions for firm a and b
plot_br(model)
# Plot second order best response function for firm a
plot_br(model,what='br2')
```
